```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
movies =pd.read_csv('data/tmdb_5000_movies.csv')
credits =pd.read_csv('data/tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
credits.head(2)
```

```python
movies = movies.merge(credits ,on ='title')
movies.head(2)
```

```python
movies.iloc[0]
```

```python
movies.columns
```

```python
movies =movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.shape
```

```python
import ast

# convert_lambda =lambda text: [i['name'] for i in ast.literal_eval(text)]

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
        
    return l

movies.head(2)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
def convert_cast(text):
    l = []
    counter =0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter +=1
        
    return l

# movies.iloc[0]['cast']
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(2)
```

```python
def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
        
    return l

movies.iloc[0]['crew']

movies['crew'] = movies['crew'].apply(fetch_director)

```

```python
movies.head(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
# إزالة المسافات
def remove_space(word):
    l =[]
    for i in word:
        l.append(i.replace(" ",""))
    return l

movies['crew'] = movies['crew'].apply(remove_space)
movies['cast'] = movies['cast'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)


movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df =movies[['movie_id','title','tags']].copy()
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
# new_df.head(3)
```

```python
new_df.iloc[0]['tags']
```

```python
# import nltk
from nltk.stem import PorterStemmer
```

```python
ps =PorterStemmer()
```

```python
# تحويل الكلمات إلى جذرها الأساسي.
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    
    return " ".join(l)
```

```python
new_df['tags'] =new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
# cv.get_feature_names_out()
```

```python
# vv =pd.DataFrame(vector)
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similary = cosine_similarity(vector)
# similary
```

```python
dd =pd.DataFrame(similary)
dd
```

```python
new_df[new_df['title']== 'Spider-Man']
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similary[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)

def recommendd(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similary[index])), reverse=True, key=lambda x: x[1])
    recommended_movies = []
    for i in distances[1:6]:
        recommended_movies.append(new_df.iloc[i[0]])
    recommended_movies = pd.DataFrame(recommended_movies)
    return recommended_movies
```

```python
recommendd('Spider-Man')
```
